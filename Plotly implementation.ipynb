# Spotify - All Time Top 2000s Mega Dataset

```python
filename = r'C:\Users\mitsy\Desktop\personal work\spotify\top-2000-tracks.csv'
spotify_df=pd.read_csv(filename, index_col=0, header=0)
spotify_df.head()
```

```python
def plot_features_target(spotify_df):
    show_projection = True
    fig = go.Figure()
    for i in spotify_df['Top Genre'].unique():
        df = spotify_df.loc[spotify_df['Top Genre'] == i]
        x = df['Beats Per Minute (BPM)']
        X = (x-x.min())/(x.max()-x.min())
        y = df['Liveness']
        Y = (y-y.min())/(y.max()-y.min())
        z = df['Speechiness']
        Z = (z-z.min())/(z.max()-z.min())
        fig.add_trace(go.Scatter3d(
            x=X,
            y=Y,
            z=Z,
            name = i,
            mode='markers',               
            marker=dict(
                size=3#,
                #color=
                ),

            hovertemplate = 
            '<b>'+'Beats Per Minute (BPM)'+'</b>: %{x:.2f}<br><extra></extra>'+
            '<b>'+'Liveness'+'</b>: %{y:.2f}<br>'+
            '<b>'+'Speechiness'+'</b>: %{z:.2f}',
            projection = dict(
                    x=dict(show=show_projection,opacity=0.5),
                    y=dict(show=show_projection,opacity=0.5),
                    z=dict(show=show_projection,opacity=0.5))
        ))

    title = dict(text='Spotify - All Time Top 2000s Mega Dataset',
                 x=0.5, 
                 y = 0.95,
                 font = dict(color = 'white',family='monospace')
                )
    scene = dict(xaxis=dict(color='white',title='Beats Per Minute (BPM)',backgroundcolor='black',showline=False,linecolor='white',titlefont=dict(color='white',family='arial')),
                yaxis=dict(color='white',title='Liveness',backgroundcolor='black',showline=False,linecolor='white',titlefont=dict(color='white',family='arial')),
                zaxis=dict(color='white',title='Speechiness',backgroundcolor='black',showline=False,linecolor='white',titlefont=dict(color='white',family='arial')
                ))

    
    fig.update_layout(showlegend = True,
                      legend = dict(font=dict(color='white'),x=0.8),
                      title=title,
                      width=1600, height=800,
                      margin= dict(l=5,r=5,b=5,t=5),
                      paper_bgcolor="black",
                      scene = scene
               )
    
    return fig
```

```python
fig = plot_features_target(spotify_df)
fig.write_html('All Time Top 2000s Mega Dataset.html')
```
